```python
import pandas as pd
import os
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
model = LinearRegression()
# print(os.getcwd())
os.chdir('C:/Users/ngocdung/Dropbox/Front-end developer/src/data-science-path/udacity-nanodegree/Project 1/Project 1')
diamonds = pd.read_csv("diamonds.csv")
new_diamonds = pd.read_csv("new-diamonds.csv")
print(diamonds.head())
print(new_diamonds.head())
```

```python
# Since cut, clarity has more meaningful ordinal altneratives (cut_ord, clarity_ord), exclude those variables from the model
x = diamonds[['carat', 'cut_ord', 'clarity_ord']]
y = diamonds['price']
model.fit(x, y)
print(x.head())
new_x = new_diamonds[['carat', 'cut_ord', 'clarity_ord']]
new_diamonds['price'] = model.predict(new_x)
print(new_diamonds.head())
```

```python
plt.scatter(diamonds['carat'], diamonds['price'])
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
plt.scatter(diamonds['carat'], diamonds['price'], label = 'Training data')
plt.scatter(new_diamonds['carat'], new_diamonds['price'], label = 'New data')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.legend()
plt.show()
```
